```python

def get_neighbours(matrix, node):
    x_min = 0
    y_min = 0
    x_max = len(matrix) - 1
    y_max = len(matrix[0]) - 1
    x, y = node

    height = ord(matrix[x][y]) - 2
    neighbours = []

    if (x_min < x) and (height < ord(matrix[x-1][y])):
        neighbours.append((x-1, y))
    if (x < x_max) and (height < ord(matrix[x+1][y])):
        neighbours.append((x+1, y))
    if (y_min < y) and (height < ord(matrix[x][y-1])):
        neighbours.append((x, y-1))
    if (y < y_max) and (height < ord(matrix[x][y+1])):
        neighbours.append((x, y+1))

    return neighbours

```

```python
def bfs(mapp, endd):
    front = [endd]
    _level = dict()
    _level[endd] = 0
    while front:
        current = front.pop(0)
        for neighbour in get_neighbours(mapp, current):
            if neighbour not in _level:
                _level[neighbour] = _level[current]+1
                front.append(neighbour)
    return _level

```

```python
start = None
end = None
mapp = []
starts = []

with open("day-12-input.txt", "r") as input_file:
    for x, line in enumerate(input_file):
        mapp.append([])
        # print(mapp)
        # input()
        for y, levell in enumerate(line.strip()):
            if levell == "S":
                mapp[x].append("a")
                start = (x,y)
                starts.append(start)
            elif levell == "E":
                mapp[x].append("z")
                end = (x,y)
            elif levell == "a":
                starts.append((x,y)) # FOR PART 2
                mapp[x].append(levell)
            else:
                mapp[x].append(levell)

l = bfs(mapp, end)
```
